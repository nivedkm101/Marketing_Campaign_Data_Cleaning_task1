```python
df = pd.read_csv('/content/marketing_campaign.csv', sep='\t')
```

```python
print(df.head())
```

```python
df.info()
df.describe()
```

```python
duplicate_count = df['ID'].duplicated().sum()
print("Duplicate IDs found:", duplicate_count)
```

```python
df=df.drop(columns=['Z_CostContact', 'Z_Revenue'])
print(df.head())
```

```python
df['Income'].fillna(df['Income'].median(), inplace=True)
```

```python
null_count_income = df['Income'].isnull().sum()
print("Null values in 'Income' column:", null_count_income)
```

```python
df['Age'] = 2025 - df['Year_Birth']
df = df.drop(columns='Year_Birth')
```

```python
df.columns = df.columns.str.lower().str.replace('[^a-z0-9_]', '', regex=True)
df.head()
```

```python
new_column_names = {
    'mntwines': 'amt_wines',
    'mntfruits': 'amt_fruits',
    'mntmeatproducts': 'amt_meatproducts',
    'mntfishproducts': 'amt_fishproducts',
    'mntsweetproducts': 'amt_sweetproducts',
    'mntgoldprods': 'amt_goldprods',

}
df = df.rename(columns= new_column_names)
```

```python
df['income'] = df['income'].astype('int64')
```

```python
df['dt_customer'] = pd.to_datetime(df['dt_customer'], format='%d-%m-%Y', errors='coerce')
```

```python
status_to_replace = ['Alone','Absurd','YOLO']
df['marital_status']= df['marital_status'].replace(status_to_replace,'Other')
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.to_csv("marketing_campaign_cleaned.csv",index=False)
```
